## Vizualização dos Dados da Fifa:

```python
#Importando as bibliotecas
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# O parse_true serve para identificar os dados como datas
# O index serve pra especificar de que colunas vamos pegar os dados
fifa_filepath = "/content/fifa.csv"
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates= True)
```

```python
# Vizualizando os Dados
fifa_data.head()
```

```python
#Escolhendo a altura e largura do  gráfico
# Plotando gráfico
plt.figure(figsize = (16,6))
sns.lineplot(data = fifa_data)
```

## Vizualização de dados de um Bar

```python
#Como nao lidamos com datas nao usamos o parse_dates
flight_filepath = "/content/flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
#Printando o data_set inteiro
flight_data
```

```python
#Criando um gráfico de barras
plt.figure(figsize = (10,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay (in minutes)")
#no eixo X contem os meses
#no eixo Y usamos a coluna "NK"
```

```python
#Criando um mapa de calor
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=flight_data, annot=True)
#O annot garante que os valores de cada célula apareçam no gráfico
plt.xlabel("Airline")
```

##Gráficos de Dispersão:

```python
insurance_filepath = "/content/insurance.csv"
insurance_data = pd.read_csv(insurance_filepath)
insurance_data.head()
```

```python
#escolhendo os parametros para o gráfico de dipersão e plotando
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#Verificar a força da relação utilizando uma linha de regressão
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#Podemos utilizar o gráfico de dispersão com mais de uma várivel
#codificando os pontos por cores
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
#Vamos adicionar duas linhas de regressão para entender melhor o comportamento
# do gráfico e sua relação
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

## Distruibuição de dados

```python

iris_filepath = "/content/iris.csv"

iris_data = pd.read_csv(iris_filepath, index_col="Id")

iris_data.head()
```

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species")
```
